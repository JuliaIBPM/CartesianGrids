# Plotting fields

```julia
using CartesianGrids
```

```julia
using Plots
pyplot()
```

### Plotting scalar fields

```julia
g = PhysicalGrid((-3,3),(-2,2),0.02)
w = Nodes(Dual,size(g));
gauss = SpatialGaussian(0.5,0,0,1)
gfield = GeneratedField(w,gauss,g);
w .= gfield();
```

#### With no x or y data

```julia
plot(w)
```

#### Manually getting the x, y coordinates

```julia
xg, yg = coordinates(w,g)
```

```julia
plot(xg,yg,w)
```

#### With x, y data taken from the grid data automatically

```julia
plot(w,g,trim=1)
```

```julia
plot(gfield)
```

### Plotting vector field

```julia
q = Edges(Dual,w);
qfield = GeneratedField(q,gauss,EmptySpatialField(),g);
q .= qfield();
```

```julia
plot(q)
```

```julia
plot(q,g)
```

```julia
plot(qfield)
```

### Plotting tensor field

```julia
dq = EdgeGradient(Dual,w);
tfield = GeneratedField(dq,[gauss,EmptySpatialField(),gauss,gauss],g);
dq .= tfield();
```

```julia
plot(dq)
```

```julia
plot(dq,g)
```

```julia
plot(tfield)
```
